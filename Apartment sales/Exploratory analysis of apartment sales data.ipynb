# Описание проекта
В нашем распоряжении данные сервиса Яндекс Недвижимость — архив объявлений за несколько лет о продаже квартир в Санкт-Петербурге и соседних населённых пунктах.

О каждой квартире в базе содержится два типа данных: добавленные пользователем и картографические. Например, к первому типу относятся площадь квартиры, её этаж и количество балконов, ко второму — расстояния до центра города, аэропорта и ближайшего парка. 

**Задача:** выполнить предобработку данных и изучить их, чтобы найти интересные особенности и зависимости, которые существуют на рынке недвижимости.

**Описание данных:**
- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м)
- `balcony` — число балконов
- `ceiling_height` — высота потолков (м)
- `cityCenters_nearest` — расстояние до центра города (м)
- `days_exposition` — сколько дней было размещено объявление (от публикации до снятия)
- `first_day_exposition` — дата публикации
- `floor` — этаж
- `floors_total` — всего этажей в доме
- `is_apartment` — апартаменты (булев тип)
- `kitchen_area` — площадь кухни в квадратных метрах (м²)
- `last_price` — цена на момент снятия с публикации
- `living_area` — жилая площадь в квадратных метрах (м²)
- `locality_name` — название населённого пункта
- `open_plan` — свободная планировка (булев тип)
- `parks_around3000` — число парков в радиусе 3 км
- `parks_nearest` — расстояние до ближайшего парка (м)
- `ponds_around3000` — число водоёмов в радиусе 3 км
- `ponds_nearest` — расстояние до ближайшего водоёма (м)
- `rooms` — число комнат
- `studio` — квартира-студия (булев тип)
- `total_area` — общая площадь квартиры в квадратных метрах (м²)
- `total_images` — число фотографий квартиры в объявлении

## Вторая часть. Предобработка данных

### Изучение и устранение пропущенных значений в столбцах

Изучим сумму пропусков в столбцах

```python
data.isna().sum()
```

В следющих столбцах есть смысл замены пропущенных значений:
- `is_apartment` - логично предположить, что пропуск означает, что данная квартира не является апартаментом. 
- `balcony` - в данном случае пропуск, может говорить о том, что в квартире нет балконов.
- `parks_around3000` и `ponds_around3000` - пропуски в этих столбцах говорят нам об отсутвии даннных объектов в заданном диапозоне.

Заменим пропуски в данных столбцах на подходящие значения.

В столбце `is_apartment` и `balcony` используются булевы значения. Заменим пропуски на `False`.

Для остальных столбцов используются числовые значения. Заменим пропуски на `0`.

```python
data['is_apartment'] = data['is_apartment'].fillna(False)
data['balcony'] = data['is_apartment'].fillna(False)
data['parks_around3000'] = data['parks_around3000'].fillna(0)
data['ponds_around3000'] = data['ponds_around3000'].fillna(0)
```

Пропуски в остальных столбцах заменять на какие либо значения будет некорректно. 

Для объяснения причины возникновения первым делом стоит учитывать человеческий фактор, и кто-то просто пропустил значение. Особенно это касается таких значений как `ceiling_height`, `floors_total`, `locality_name`.

Говоря о `kitchen_area` и `living_area`, то кухня в квартире может отсутствовать, а квартиру-студию порой тяжело разделить на жилую зону и кухню.

Пропуски в столбцах с информацией об удаленности каких либо объектов (`airports_nearest`, `cityCenters_nearest`, `parks_nearest`, `ponds_nearest`) могут говорить о том, что рядом в принципе нету этих объектов, либо находятся они так далеко, что указывать их смысла нет.

А вот судя по пропускам в `days_exposition` можно сделать вывод, что на момент получения данных эти квартиры еще не были проданы.

### Преобразование типов данных

Рассмотрим типы данных в каждом столбце и определим в каких нужно изменить тип данных.

Выведим общую информацию.

```python
data.info()
```

Для удобства работы и удобного отображения результатов, заменим тип данных в следующих столбцах на тип данных `int64`:
- `last_price`
- `parks_around3000`
- `ponds_around3000`

А для дальнейшей работы с датой заменим тип данных столбца `first_day_exposition` на `datetime`, а `locality_name` на `str`.


```python
data['last_price'] = data['last_price'].astype('int')
data['parks_around3000'] = data['parks_around3000'].astype('int')
data['ponds_around3000'] = data['ponds_around3000'].astype('int')
data['locality_name'] = data['locality_name'].astype('str')
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], 
                                              format='%Y-%m-%dT%H:%M:%S') #Формат - "Год-Месяц-День Время"
```

Посмотрим на уникальные значения столбца `locality_name`

```python
data['locality_name'].unique()
```

Видно, что в столбце есть неявные дублкаты "поселок" и "посёлок".

## Третья часть. Добавление новых столбцов

Добавим в таблицу следующие столбцы:
- цена одного квадратного метра;
- день недели публикации объявления (0 — понедельник, 1 — вторник и так далее);
- месяц публикации объявления;
- год публикации объявления;
- тип этажа квартиры (значения — «первый», «последний», «другой»);
- расстояние до центра города в километрах (переведем из м в км и округлим до целых значений).    

```python
data['square_price'] = data['last_price'] / data['total_area'] # цена 1 кв.м. - цена деленая на общую площадь

data['publication_day'] = data['first_day_exposition'].dt.weekday # столбец с днем недели публикации

data['publication_month'] = data['first_day_exposition'].dt.month # столбец с месяцем публикации

data['publication_year'] = data['first_day_exposition'].dt.year # столбец с годом публикации

data['cityCenters_nearest_km'] = round(data['cityCenters_nearest']/1000) # столбец расстоянием до центра города в км
```

```python
def type_floor (row): # функция для присваивания типа этажа
    floor = row['floor']
    total_floors = row['floors_total']
    if floor == total_floors:
        return 'последний'
    elif floor == 1:
        return 'первый'
    else:
        return 'другой'
    
data['floor_type'] = data.apply(type_floor, axis=1) # столбец с типом этажа    
```

### Изучение факторов влияющих на общую (полную) стоимость объекта.

Изучим, зависит ли цена от:
- общей площади;
- жилой площади;
- площади кухни;
- количества комнат;
- этажа, на котором расположена квартира (первый, последний, другой);
- даты размещения (день недели, месяц, год).

Построим графики, которые покажут зависимость цены от указанных выше параметров.

Для начала создадим отдельную переменную, в которой избавимся от "аномальных значений" цены.

```python
clear_data = data.copy()
clear_data['last_price_mln'] = clear_data['last_price']/1000000 # переведем для удобства цены в миллионы

clear_data.plot(x='last_price_mln', y='days_exposition', kind='scatter', alpha=0.3, figsize=(5,3))
plt.suptitle("Распределение квартир по сроку продажи")
plt.xlabel("Цена, млн")
plt.ylabel("Дней в продаже")
```

Видно, что объявлений с ценой больше 100 млн еденичны. Удалим такие объявления.

```python
clear_data = clear_data[clear_data['last_price_mln']<100]
```

**Завизимость Общей площади**

```python
print('Коэффициент корреляции цены и общей площади квартиры:', clear_data['last_price_mln'].corr(clear_data['total_area']))
clear_data.plot(x='last_price_mln', y='total_area', kind='scatter', alpha=0.05, figsize=(5,3))
plt.suptitle("Распределение стоимости квартир по общей площади")
plt.xlabel("Цена")
plt.ylabel("Площадь, м²")
```

К.ф. корреляции очень высокий, можно утверждать, что увеличение общей площади значительно влияет на цену квартиры.

**Завизимость Жилой зоны**

```python
print('Коэффициент корреляции цены и жилой площади:', clear_data['last_price_mln'].corr(clear_data['living_area']))
clear_data.plot(x='last_price_mln', y='living_area', kind='scatter', alpha=0.3, figsize=(5,3))
plt.suptitle("Распределение квартир по жилой площади")
plt.xlabel("Цена")
plt.ylabel("Жилая площадь, м2")
```

К.ф. корреляции меньше, но тоже достаточно высокий, чтобы утверждать, что увеличение жилой площади влияет на цену квартиры.

**Завизимость кухонной зоны**

```python
print('Коэффициент корреляции цены и площади кухни:', clear_data['last_price_mln'].corr(clear_data['kitchen_area']))
clear_data.plot(x='last_price_mln', y='kitchen_area', kind='scatter', alpha=0.3, figsize=(5,3))
plt.suptitle("Распределение квартир по площади кухни")
plt.xlabel("Цена, млн")
plt.ylabel("Площадь, м²")
```

К.ф. корреляции в районе средних значений, видно, что вляет на цену, но по сравнению с другими площадями - меньше всех. 

**Завизимость кол-ва комнат**

```python
print('Коэффициент корреляции цены и кол-ва комнат:', clear_data['last_price_mln'].corr(clear_data['rooms'])) 
clear_data.sort_values('last_price_mln').plot(x='last_price_mln', y='rooms', kind='scatter', alpha=0.3, figsize=(5,3))
plt.suptitle("Распределение квартир по кол-ву комнат")
plt.xlabel("Цена, млн")
plt.ylabel("Кол-во комнат")
```

К.ф. корреляции в районе средних значений. Но на графике видно, что много квартир, с кол-ом комнат гараздо большим, чем "обычно" есть в квартирах. Попробуем постепенно уменьшить кол-во комнат датафрейме. 

```python
clear_room_data = clear_data.loc[clear_data['rooms']<=5]
print('Коэффициент корреляции цены и кол-ва комнат:', clear_room_data['last_price_mln'].corr(clear_room_data['rooms']))
clear_room_data = clear_data.loc[clear_data['rooms']<=4]
print('Коэффициент корреляции цены и кол-ва комнат:', clear_room_data['last_price_mln'].corr(clear_room_data['rooms']))
clear_room_data = clear_data.loc[clear_data['rooms']<=3]
print('Коэффициент корреляции цены и кол-ва комнат:', clear_room_data['last_price_mln'].corr(clear_room_data['rooms']))
```

К.ф. корреляции падает при уменьшении кол-ва комнат.

```python
clear_room_data['last_price_mln'].corr(clear_data['total_area'])
```

А вот к.ф. корреляции общей площади так-же высок.

**Завизимость от даты размещения**

```python
clear_year_pivot = clear_data.pivot_table(index = 'publication_year', values = 'last_price_mln', 
                                          aggfunc = ['mean', 'count', 'median'])
clear_year_pivot.columns = ['mean', 'count', 'median']
clear_year_pivot.plot(y = 'median', figsize=(5,3))
plt.suptitle("Распределение медианной цены по годам")
plt.xlabel("Год публикации")
plt.ylabel("Цена, млн")
clear_year_pivot
```

Заметно, что цены на квартиры стремительно падали с 14 по 16 год. Рост цен на квартиры начался только с 2018 года.

```python
clear_month_pivot = clear_data.pivot_table(index = 'publication_month', values = 'last_price_mln', aggfunc = ['mean', 'count', 'median'])
clear_month_pivot.columns = ['mean', 'count', 'median']
clear_month_pivot.plot(y = 'median', figsize=(5,3))
plt.suptitle("Распределение медианной цены по месяцам")
plt.xlabel("Месяц публикации")
plt.ylabel("Цена, млн")
clear_month_pivot
```

Заметно, что апрельские объявления  - объявления с самой высокой ценой. После чего идет резкий спад, который достигает своих минимальных значений в июне. Дальше идет постепенный рост и к июлю цена постепенно выравнивается.

```python
clear_day_pivot = clear_data.pivot_table(index = 'publication_day', values = 'last_price_mln', aggfunc = ['mean', 'count', 'median'])
clear_day_pivot.columns = ['mean', 'count', 'median']
clear_day_pivot.plot(y = 'median', figsize=(5,3))
plt.suptitle("Распределение медианной цены по дням недели")
plt.xlabel("День недели публикации")
plt.ylabel("Цена, млн")
clear_day_pivot
```

Во вторник и среду, цены на квартиры самые высокие. В выходные самые низкие.
